```python


for key, x in list(monthly_indicators.items()):
    x['name']= key
    column_headers = list(x.columns.values)
    del column_headers[-1]
    del column_headers[0:4]
    if (('siec' not in x.columns) and ('nrg_bal' not in x.columns)):
        temp_value = x.melt(id_vars=['geo\\time',"unit","name"], var_name='time', value_vars=column_headers)
        temp_value = time_change(temp_value)
        temp_value = change_unit(temp_value)
        temp_value = temp_value.pivot_table(values= ["value"],index= ["geo\\time","year","month","time"], columns= ["unit"],aggfunc=np.sum)
        
    elif (('siec' in x.columns) and ('nrg_bal' not in x.columns)):
        temp_value = x.melt(id_vars=['geo\\time',"unit","siec","name"], var_name='time', value_vars=column_headers)
        temp_value = time_change(temp_value)
        temp_value = change_siec(temp_value) 
        temp_value = change_unit(temp_value)
        temp_value = temp_value.pivot_table(values= ["value"],index= ["geo\\time","year",'unit',"month","time"], columns= ["siec"],aggfunc=np.sum)
    
    elif (('siec' not in x.columns) and ('nrg_bal' in x.columns)):
        temp_value = x.melt(id_vars=['geo\\time',"unit","nrg_bal","name"], var_name='time', value_vars=column_headers) 
        temp_value = time_change(temp_value)   
        temp_value = change_en_bal(temp_value)
        temp_value = change_unit(temp_value)
        temp_value = temp_value.pivot_table(values= ["value"],index= ["geo\\time","year",'unit',"month","time"], columns= ["nrg_bal"],aggfunc=np.sum)
        
    else:
        temp_value = x.melt(id_vars=['geo\\time',"unit","siec","nrg_bal","name"], var_name='time', value_vars=column_headers)
        temp_value = time_change(temp_value)
        temp_value = change_siec(temp_value)
        temp_value = change_en_bal(temp_value)
        temp_value = change_unit(temp_value)
        temp_value = temp_value.pivot_table(values= ["value"],index= ["geo\\time","year",'unit',"month","time"], columns= ["nrg_bal","siec"],aggfunc=np.sum)
    
    temp_value = temp_value.reset_index()
    temp_value.columns = temp_value.columns.map('|'.join).str.strip('|')
    
    monthly_indicators[key] = temp_value
    
  
```

```python
import dash, dash_table
import dash_core_components as dcc
import dash_html_components as html
import dash_bootstrap_components as dbc
import pandas as pd
import plotly.graph_objs as go
from dash import Input, Output, dcc, html
from sklearn import datasets
from sklearn.cluster import KMeans

import dash_table as dt
from dash_extensions import Download
import dash_table_experiments as dte
from flask import send_file
import io
import flask


app = dash.Dash(external_stylesheets=[dbc.themes.BOOTSTRAP])


    

colors = {
    'background': '#111111',
    'text': '#7FDBFF'
}


# --------------------------------------------------------------------- TAB 1 - Energy - Start preparation code
ene_country_options = []
for country in energy_per_product['geo\time'].unique():
    ene_country_options.append({'label':str(country),'value':country})


ene_columns_selector = []

for column in energy_per_product.columns:
    ene_columns_selector.append(column)

ene_columns_default_values = ene_columns_selector[0:5]



ene_features = energy_per_product.columns












starting_df = monthly_indicators["Supply_and_transformation_of_solid_fossil_fuels"]



monthly_country_options = []
for x in starting_df['geo\\time'].unique():
    monthly_country_options.append({'label':str(x),'value':x})


dataframe_available = list(monthly_indicators.keys())



colors = {
    'background': '#111111',
    'text': '#7FDBFF'
}

msel_features = starting_df.columns


# --------------------------------------------------------------------- TAB 1 - Energy - End preparation code




# --------------------------------------------------------------------- TAB 2 - Economy - Start preparation code

eco_country_options = []
for country in economy['geo\time'].unique():
    eco_country_options.append({'label':str(country),'value':country})

eco_unit_options = []
for unit in economy['unit'].unique():
    eco_unit_options.append({'label':str(unit),'value':unit})

eco_features = economy.columns


# --------------------------------------------------------------------- TAB 2 - Economy - End preparation code





# --------------------------------------------------------------------- TAB 3 - Gas - Start preparation code

gas_country_options = []
for country in gaschart2['geo\time'].unique():
    gas_country_options.append({'label':str(country),'value':country})

gas_month_options = []
for month in gaschart1['Month'].unique():
    gas_month_options.append({'label':str(month),'value':month})
    

# --------------------------------------------------------------------- TAB 3 - Gas - End preparation code






# --------------------------------------------------------------------- TAB 4 - Clusters - Start preparation code



clu_country_options = []
for country in chart_emissions['geo\time'].unique():
    clu_country_options.append({'label':str(country),'value':country})



controls = html.Div(
    [
        html.Div(
            [
                dbc.Label("X variable"),
                dcc.Dropdown(
                    id="x-variable",
                    options=[
                        {"label": col, "value": col} for col in clust_df.columns
                        
                    ],
                    value="sepal length (cm)",
                ),
            ], className="simple_box dropdown_box four columns"),
        
        html.Div(
            [
                dbc.Label("Y variable"),
                dcc.Dropdown(
                    id="y-variable",
                    options=[
                        {"label": col, "value": col} for col in clust_df.columns
                    ],
                    value="sepal width (cm)",
                ),
            ], className="simple_box dropdown_box four columns"),
        
        html.Div(
            [
                dbc.Label("Cluster count"),
                dbc.Input(id="cluster-count", type="number", value=3),
            ], className="simple_box dropdown_box four columns"),
    ], className="row flex-display")



# --------------------------------------------------------------------- TAB 4 - Clusters - End preparation code








# ------------------------------------ APP AND TABS - Start ---------------------------------------
app.layout= html.Div(children = [
    


html.Div([
   

    
    
  html.Div([  
  html.Div([html.H1('TM470')],className="ds_title"),
  html.Div([html.H2('Eurostat data dashboard')],className="ds_subtitle")]
  , className = "ds_titles_container"),  
    
  html.Div([
    html.Div([
        dcc.Tabs(id = "tabs-styled-with-inline", value = 'tab-1', children = [
            dcc.Tab(label = 'Energy Indicators', value = 'tab-1'),
            dcc.Tab(label = 'Economy Indicators', value = 'tab-2'),
            dcc.Tab(label = 'Gas Indicators', value = 'tab-3'),
            dcc.Tab(label = 'Greenhouse Emissions Clustering', value = 'tab-4'),
        ], className = "tabs_container"),
        html.Div([
            
            
            
        ], id = 'tabs-content-inline', className="tab-main-class")
    ], className = "create_container3 twelve columns", ),
    ], className = "row flex-display"),
    
    
    
   
    

    
]),
    
    html.Div([
        html.A(['Datasource from Eurostat'], href="https://ec.europa.eu/eurostat/data/database", target="_blank"),
        html.Span('Eurostat is not responsable for the development of the dashboard')
    ], className = "row flex-display footer"),


])

# ------------------------------------ APP AND TABS - End ---------------------------------------



def tab1_table(columns_to_display):
    return dash_table.DataTable(
        id='datatable-interactivity',
        columns=[
            {"name": i, "id": i, "deletable": True, "selectable": True} for i in columns_to_display
        ],
        data=energy_per_product.to_dict('records'),
        editable=True,
        filter_action="native",
        sort_action="native",
        sort_mode="multi",
        column_selectable="single",
        row_selectable="multi",
        row_deletable=True,
        selected_columns=[],
        selected_rows=[],
        page_action="native",
        page_current= 0,
        page_size= 10,
        )
    


def tab1_function():
    return html.Div(children = [
    
        
        html.Div([html.H3('Multi variable chart')], className="simple_title"),
         
        html.Div([
        html.Div([
            html.Div([html.Label('Select Total Energy Supply or Final Energy Consumption')], className="simple_box_header"),
            dcc.Dropdown(
                id='x1axis',
                options=[{'label': i.title(), 'value': i} for i in ene_features],
                value='displacement'
            )
        ], className="simple_box dropdown_box four columns"),

        html.Div([
            html.Div([html.Label('Select Total Energy Supply or Final Energy Consumption')], className="simple_box_header"),
            dcc.Dropdown(
                id='x2axis',
                options=[{'label': i.title(), 'value': i} for i in ene_features],
                value='acceleration'
            )
        ], className="simple_box dropdown_box four columns"),
        
        html.Div([
            html.Div([html.Label('Select Country')], className="simple_box_header"),
            dcc.Dropdown(id='country-picker',options=ene_country_options,value= 'EU27_2020')
        ], className="simple_box dropdown_box four columns")], className = "row flex-display"),
    
    
    html.Div([    
    dcc.Graph(id='graph')
    
    ],className="simple_box simple_graph row flex-display"),
    
    html.Div(className="simple_separator"),
        
        
    html.Div([html.H3('Dataframe table')], className="simple_title"),
        
    html.Div([     
        html.Div([
            html.Div([html.Label('Select columns to display')], className="simple_box_header"),
            dcc.Dropdown(id='columns_selector',options=ene_columns_selector,multi=True,value=ene_columns_default_values)]),
            ], className="simple_box dropdown_box row flex-display"),
    html.Div([ 
        html.Div([tab1_table(ene_columns_default_values)],id="table-wrapper",className="energy-table-wrapper"),
        html.Div(id='datatable-interactivity-container')
    ], className="simple_box simple_graph row flex-display"),
    
    
    html.Div(className="simple_separator"),
        
    html.Div([ 
    
    html.Div([
            html.Div([html.Label('Select Dataframe')], className="simple_box_header"),
            dcc.Dropdown(
                id='df-picker1',
                options=dataframe_available,value= 'Supply_and_transformation_of_solid_fossil_fuels'
            )
        ], className="simple_box dropdown_box four columns"),
    
    html.Div([
            html.Div([html.Label('Select Country')], className="simple_box_header"),
            dcc.Dropdown(id='msel_country-picker',options=monthly_country_options,value= 'EU27_2020')
        ], className="simple_box dropdown_box four columns")
    
    
    ], className="row flex-display"),
     
    html.Div([    
    html.Div([
            html.Div([html.Label('Select attribute 1')], className="simple_box_header"),
            dcc.Dropdown(
                id='msel_x1axis',
                options=[{'label': i.title(), 'value': i} for i in msel_features],
                value='displacement'
            )
        ], className="simple_box dropdown_box four columns"),
        
    html.Div([
            html.Div([html.Label('Select attribute 2')], className="simple_box_header"),
            dcc.Dropdown(
                id='msel_x2axis',
                options=[{'label': i.title(), 'value': i} for i in msel_features],
                value='displacement'
            )
        ], className="simple_box dropdown_box four columns"),
    
    ], className="row flex-display"),

        
    html.Div([   
    html.Div([    
    dcc.Graph(id='msel_graph')
    
    ],className="simple_box simple_graph row flex-display")
    ], className="row flex-display")    
    
])



def tab2_function():
    return html.Div(children = [
    
    
        html.Div([html.H3('Economic indicators chart')], className="simple_title"),
    
        html.Div([
        html.Div([
            html.Div([html.Label('Select variable (unit)')], className="simple_box_header"),
            dcc.Dropdown(
                id='x1axis',
                options=[{'label': i.title(), 'value': i} for i in eco_features],
                value='displacement'
            )
        ], className="simple_box dropdown_box four columns"),

        html.Div([
            html.Div([html.Label('Select country')], className="simple_box_header"),
            dcc.Dropdown(id='eco_country-picker',options=eco_country_options,value= 'EU27_2020')
        ], className="simple_box dropdown_box four columns"),
        
        html.Div([
            html.Div([html.Label('Select unit')], className="simple_box_header"),
            dcc.Dropdown(id='eco_unit-picker',options=eco_unit_options,value= 'CLV10_EUR_HAB')
        ], className="simple_box dropdown_box four columns"),
        
        ], className = "row flex-display"),
    
    html.Div([
    
    dcc.Graph(id='eco_graph'),
    ], className="simple_box simple_graph row flex-display"),
        
        
    html.Div(className="simple_separator"),
        
        
    html.Div([html.H3('Economic indicators table')], className="simple_title"),

    html.Div([
    dash_table.DataTable(
        id='eco_datatable-interactivity',
        columns=[
            {"name": i, "id": i, "deletable": True, "selectable": True} for i in economy.columns
        ],
        data=economy.to_dict('records'),
        editable=True,
        filter_action="native",
        sort_action="native",
        sort_mode="multi",
        column_selectable="single",
        row_selectable="multi",
        row_deletable=True,
        selected_columns=[],
        selected_rows=[],
        page_action="native",
        page_current= 0,
        page_size= 10,
    )], className="simple_box simple_graph row flex-display"),
    html.Div(id='eco_datatable-interactivity-container')
])






def tab3_function():
    return html.Div(children = [
    
    
    html.Div([html.H3('Inland consumption of Gas in 2022')], className="simple_title"),
    html.Div([
    html.Div([html.Label('Select month')], className="simple_box_header"),
    dcc.Dropdown(id='gas_month-picker',options=gas_month_options,value= '01')], className="simple_box dropdown_box row flex-display"),
    html.Div([dcc.Graph(id='gas_graph4')], className="simple_box simple_graph row flex-display"),
            
    html.Div(className="simple_separator"),
        
        
    html.Div([html.H3('Gas import and exports')], className="simple_title"),
    html.Div([
    html.Div([html.Label('Select month')], className="simple_box_header"),    
    dcc.Dropdown(id='gas_monthly-picker',options=gas_month_options,value= '01')], className="simple_box dropdown_box row flex-display"),
    html.Div([dcc.Graph(id='gas_graph3')], className="simple_box simple_graph row flex-display"),
    
])






def tab4_function():
    return html.Div(children = [
        html.Div([html.H3('K-means clusterings')], className="simple_title"),
        html.Div([html.H5('Visualize clustering of EU Countries by choosing the variables in the dropdown menu')], className="simple_subtitle"),
        
        
        controls,
        html.Div([dcc.Graph(id="cluster-graph")], className="simple_box simple_graph row flex-display"),

        html.Div(className="simple_separator"),  
        
        html.Div([html.H3('Gas emissions')], className="simple_title"),
        html.Div([
            html.Div([html.Label('Select country')], className="simple_box_header"),
            dcc.Dropdown(id='country-picker',options=clu_country_options,value= 'EU27_2020')
            ], className="simple_box dropdown_box row flex-display"),
        html.Div([dcc.Graph(id='graph3')], className="simple_box simple_graph row flex-display"),
        
        
        
    ]
        
    #fluid=True
)





# ---------------------------------- Tabs management - Start -----------------------------------------




@app.callback(Output('tabs-content-inline', 'children'),
              Input('tabs-styled-with-inline', 'value'))

def update_tab(tab):
    if tab == 'tab-1':
        return tab1_function()
    elif tab == 'tab-2':
        return tab2_function()
    elif tab == 'tab-3':
        return tab3_function()
    elif tab == 'tab-4':
        return tab4_function()
# ---------------------------------- Tabs management - End -----------------------------------------






# ---------------------------------- Tab 1 - Start -----------------------------------------
@app.callback(    [
    Output('msel_x1axis', 'options'),
    Output('msel_x2axis', 'options')

    
    ],
    [
    Input(component_id='df-picker1', component_property='value')
    
    ]
)


def update_dropdown(dropdown_input):
    global starting_df
    starting_df = monthly_indicators[dropdown_input]
    msel_features = starting_df.columns

    return starting_df.columns,starting_df.columns



@app.callback(
    Output('msel_graph', 'figure'),
    [
    Input(component_id='msel_country-picker', component_property='value'),
        
    Input('msel_x1axis', 'value'),
    Input('msel_x2axis', 'value')
    
    ]
)

  
    
def msel_update_figure(selected_country,msel_x1axis,msel_x2axis):
    filtered_df = starting_df[starting_df['geo\\time'] == selected_country]
    
   
   
    trace4 =go.Bar(x= filtered_df['year'], y = filtered_df[msel_x1axis], name = msel_x1axis, marker = {'color' : '#bbbbbb'})
    trace5 =go.Bar(x= filtered_df['year'], y = filtered_df[msel_x2axis], name = msel_x2axis, marker = {'color' : '#77cc77'})
    data = [trace4,trace5]
    layout = go.Layout(title = 'Energy')
    msel_fig1 = go.Figure(data = data, layout = layout)
        

    return msel_fig1
    
    
    
    

    
    


@app.callback(
    Output('graph', 'figure'),
    [Input(component_id='country-picker', component_property='value'),
    
    Input('x1axis', 'value'),
    Input('x2axis', 'value')
    
    ]
)

  
    
def update_figure(selected_country,x1axis,x2axis):
    filtered_df = energy_per_product[energy_per_product['geo\time'] == selected_country]
    
   
   
    trace4 =go.Bar(x= filtered_df['year'], y = filtered_df[x1axis], name = x1axis, marker = {'color' : '#bbbbbb'})
    trace5 =go.Bar(x= filtered_df['year'], y = filtered_df[x2axis], name = x2axis, marker = {'color' : '#77cc77'})
    data = [trace4,trace5]
    layout = go.Layout(title = 'Energy')
    fig2 = go.Figure(data = data, layout = layout)
        

    return fig2

# data table

@app.callback(
    Output('datatable-interactivity', 'style_data_conditional'),
    Input('datatable-interactivity', 'selected_columns')
)
def update_styles(selected_columns):
    return [{
        'if': { 'column_id': i },
        'background_color': '#D2F3FF'
    } for i in selected_columns]

@app.callback(
    Output('datatable-interactivity-container', "children"),
    Input('datatable-interactivity', "derived_virtual_data"),
    Input('datatable-interactivity', "derived_virtual_selected_rows"))
def update_graphs(rows, derived_virtual_selected_rows):
    
    if derived_virtual_selected_rows is None:
        derived_virtual_selected_rows = []

    dff = df if rows is None else pd.DataFrame(rows)

    colors = ['#7FDBFF' if i in derived_virtual_selected_rows else '#0074D9'
              for i in range(len(dff))]

    return [
        dcc.Graph(
            id=column,
            figure={
                "data": [
                    {
                        "x": dff["country"],
                        "y": dff[column],
                        "type": "bar",
                        "marker": {"color": colors},
                    }
                ],
                "layout": {
                    "xaxis": {"automargin": True},
                    "yaxis": {
                        "automargin": True,
                        "title": {"text": column}
                    },
                    "height": 250,
                    "margin": {"t": 10, "l": 10, "r": 10},
                },
            },
        )
        
        for column in ["total energy supply solid fossil fuel",                   
" total energy supply manufactured gas", 
"total energy supply Electricity",
"total energy supply Natural gas",
"total energy supply Heat",
"total energy supply Nuclear heat" ,
 "total energy supply Oil and petroleum (no biofuel)",       
 "total energy supply Peat",                            
 "total energy supply Renewables and biofuel",             
 "total energy supply Oil shale and sands",                  
 "total energy supply Total" ,                         
 "total energy supply Non renewable waste",                
 "unit_x",                                              
 "final energy consumption solid fossil fuel",             
 "final energy consumption manufactured gas",               
 "final energy consumption Electricity",                    
 "final energy consumption Natural gas",                  
 "final energy consumption Heat" ,                      
 "final energy consumption Nuclear heat",               
 "final energy consumption Oil and petroleum (no biofuel)",
 "final energy consumption Peat" ,                 
 "final energy consumption Renewables and biofuel",        
 "final energy consumption Oil shale and sands",            
 "final energy consumption Total",                        
 "final energy consumption Non renewable waste",
    "year", "geo\time"] if column in dff
    ]




@app.callback(
        Output('table-wrapper', 'children'),
        [Input('columns_selector', 'value')])
def update_table_live(columns_selection):
        return tab1_table(columns_selection)

# ---------------------------------- Tab 1 - End -----------------------------------------


# ---------------------------------- Tab 2 - Start -----------------------------------------



@app.callback(
    Output('eco_graph', 'figure'),
    [Input(component_id='eco_country-picker', component_property='value'),
     Input(component_id='eco_unit-picker', component_property='value'),
    
    Input('x1axis', 'value')
    
    ]
)

  
    
def eco_update_figure(selected_country,selected_unit,x1axis):
    filtered_df = economy[(economy['geo\time'] == selected_country) & (economy['unit'] == selected_unit)]
    
    
    
   
   
    trace4 =go.Scatter(x= filtered_df['year'], y = filtered_df[x1axis], name = x1axis, marker = {'color' : '#bbbbbb'})
    
    data = trace4
    layout = go.Layout(title = 'Economy')
    fig2 = go.Figure(data = data, layout = layout)
        

    return fig2

# data table

@app.callback(
    Output('eco_datatable-interactivity', 'style_data_conditional'),
    Input('eco_datatable-interactivity', 'selected_columns')
)
def eco_update_styles(selected_columns):
    return [{
        'if': { 'column_id': i },
        'background_color': '#D2F3FF'
    } for i in selected_columns]

@app.callback(
    Output('eco_datatable-interactivity-container', "children"),
    Input('eco_datatable-interactivity', "derived_virtual_data"),
    Input('eco_datatable-interactivity', "derived_virtual_selected_rows"))
def eco_update_graphs_eco(rows, derived_virtual_selected_rows):
    
    if derived_virtual_selected_rows is None:
        derived_virtual_selected_rows = []

    dff = df if rows is None else pd.DataFrame(rows)

    colors = ['#7FDBFF' if i in derived_virtual_selected_rows else '#0074D9'
              for i in range(len(dff))]

    return [
        dcc.Graph(
            id=column,
            figure={
                "data": [
                    {
                        "x": dff["country"],
                        "y": dff[column],
                        "type": "bar",
                        "marker": {"color": colors},
                    }
                ],
                "layout": {
                    "xaxis": {"automargin": True},
                    "yaxis": {
                        "automargin": True,
                        "title": {"text": column}
                    },
                    "height": 250,
                    "margin": {"t": 10, "l": 10, "r": 10},
                },
            },
        )
       
        for column in ['geo\time', 'year', 'unit',"Environmental_protection_investments_of_total_economy ('MIO_EUR', 'PC_GDP')","Gross_domestic_product_at_market_prices('CP_EUR_HAB', 'CP_MEUR')",
                       "Real_GDP_per_capita('CLV10_EUR_HAB', 'CLV_PCH_PRE_HAB')"] if column in dff
    ]



# ---------------------------------- Tab 2 - End ----------------------------------------- 

# ---------------------------------- Tab 3 - Start -----------------------------------------





@app.callback(
    Output('gas_graph4', 'figure'),
    [Input(component_id='gas_month-picker', component_property='value')
     
    ]
)


    
def gas_create_map(selected_month):
    df = gaschart1[gaschart1['Month'] == selected_month]
    radio_value='Inland consumption - observed'
    lm = None
    proj = "robinson"
    
    fig = go.Figure()
    fig.add_choropleth(
        locations=gaschart1["alpha3"],
        z=df[radio_value],
        zmin=0,
        locationmode=lm,
        colorscale="orrd",
        marker_line_width=0.5,
       
        colorbar=dict(len=0.6, x=1, y=0.5),
    )
    fig.update_layout(
        geo={
            "lataxis": {"range": [-50, 68]},
            "lonaxis": {"range": [-130, 150]},
            "projection": {"type": proj},
            "showframe": False,
        },
        margin={"t": 0, "l": 10, "r": 10, "b": 0},
    )
    return fig


@app.callback(
    Output('gas_graph3', 'figure'),
    [Input(component_id='gas_monthly-picker', component_property='value')
     
    ]
)

  
    
def gas_update_figure(selected_month):
    filtered_df = gaschart2[gaschart2['Month'] == selected_month]
    
    
    
   
   
    trace4 =go.Bar(x= filtered_df['geo\time'], y = filtered_df['Exports'], name = 'Exports', marker = {'color' : '#bbbbbb'})
    trace5 =go.Bar(x= filtered_df['geo\time'], y = filtered_df['Imports'], name = 'Imports', marker = {'color' : '#77cc77'})
    data = [trace4,trace5]
    layout = go.Layout(title = 'Gas import and export')
    fig2 = go.Figure(data = data, layout = layout)
        

        

    return fig2





# ---------------------------------- Tab 3 - End ----------------------------------------- 





# ---------------------------------- Tab 4 - Start -----------------------------------------



@app.callback(
    Output('graph3', 'figure'),
    [Input(component_id='country-picker', component_property='value')
     
    ]
)

  
    
def clu_update_figure(selected_country):
    filtered_df = chart_emissions[chart_emissions['geo\time'] == selected_country]
    
    
    
   
   
    trace4 =go.Scatter(x= filtered_df['year'], y = filtered_df['net gas emissions TOTXMEMONIA'], marker = {'color' : '#bbbbbb'})
    
    data = trace4
    layout = go.Layout(title = 'Gas emissions')
    fig2 = go.Figure(data = data, layout = layout)
        

    return fig2


@app.callback(
    Output("cluster-graph", "figure"),
    [
        Input("x-variable", "value"),
        Input("y-variable", "value"),
        Input("cluster-count", "value"),
    ],
)
def clu_make_graph(x, y, n_clusters):
    # minimal input validation, make sure there's at least one cluster
    km = KMeans(n_clusters=max(n_clusters, 1))
    df = clust_df.loc[:, [x, y]]
    km.fit(df.values)
    df["cluster"] = km.labels_

    centers = km.cluster_centers_

    data = [
        go.Scatter(
            x=df.loc[df.cluster == c, x],
            y=df.loc[df.cluster == c, y],
            
            mode='markers',
           
            marker={"size": 8},
            name="Cluster {}".format(c),
        )
        for c in range(n_clusters)
    ]

    data.append(
        go.Scatter(
            x=centers[:, 0],
            y=centers[:, 1],
            text=clust_df['geo\time'],
            textposition='top right',
            textfont=dict(color='#E58606'),
            mode='lines+markers+text',
            
           
            marker={"color": "#000", "size": 12, "symbol": "diamond"},
            name="Cluster centers",
        )
    )

    layout = {"xaxis": {"title": x}, "yaxis": {"title": y}}

    return go.Figure(data=data, layout=layout)


# make sure that x and y values can't be the same variable
def clu_filter_options(v):
    """Disable option v"""
    return [
        {"label": col, "value": col, "disabled": col == v}
        for col in clust_df.columns
    ]


# functionality is the same for both dropdowns, so we reuse filter_options
app.callback(Output("x-variable", "options"), [Input("y-variable", "value")])(
    clu_filter_options
)
app.callback(Output("y-variable", "options"), [Input("x-variable", "value")])(
    clu_filter_options
)

# ---------------------------------- Tab 4 - End -----------------------------------------








if __name__ == "__main__":
    app.run_server(debug=False)
```
